```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\nikhi\Downloads\pokemon_data.csv')
```

```python
print(df)
```

```python
df['Total']=df['HP']+df['Attack']+df["Defense"]+df['Sp. Atk']+df['Sp. Def']+df['Speed']
df
```

```python
plt.plot(df['Name'],df['Total'])
plt.show()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, # No regression line
           hue='Generation')   #hue=provides 3d dim data
```

```python
sns.boxplot(data=df)
```

```python
plt.figure(figsize=(15,6))
DF=df.drop(['Generation','Legendary','Type 1','Type 2'],axis=1)   #axis=1 defines vertically drop
sns.boxplot(data=DF)
```

```python
# Set theme
plt.figure(figsize=(15,6))
sns.set_style('darkgrid')  # or whitegrid
 
# Violin plot
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
plt.figure(figsize=(15,6))
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
# Swarm plot with Pokemon color palette
sns.swarmplot(x='Type 1', y='Attack', data=df, 
              palette=pkmn_type_colors)
```

```python
# Set figure size with matplotlib
plt.figure(figsize=(15,6))
 
# Create plot
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Attack by Type')
```

```python
# Melt DataFrame
melted_df = pd.melt(df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name="Stat") # Name of melted variable
melted_df.head()
```
